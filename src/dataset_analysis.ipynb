```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
diabetes_data = pd.read_csv('../diabetes.csv')
diabetes_data.head()
```

1. **SBP (Systolic Blood Pressure)**
  - Definition: Systolic Blood Pressure is the pressure exerted on the walls of your arteries when your heart beats and pumps blood. It’s the top number in a blood pressure reading (e.g., 120 in a reading of 120/80 mmHg).
  - Significance: High SBP (hypertension) can increase the risk of cardiovascular diseases, including heart attacks and strokes.
2. **DBP (Diastolic Blood Pressure)**
  - Definition: Diastolic Blood Pressure is the pressure in your arteries when your heart is at rest between beats. It’s the bottom number in a blood pressure reading (e.g., 80 in a reading of 120/80 mmHg).
  - Significance: High DBP is also a risk factor for heart disease. A healthy diastolic pressure is usually below 80 mmHg.
3. **FPG (Fasting Plasma Glucose)**
  - Definition: Fasting Plasma Glucose is the level of glucose (sugar) in the blood after an individual has fasted for at least 8 hours.
  - Significance: FPG is a common test for diabetes diagnosis. High FPG levels (above 100 mg/dL) may indicate prediabetes or diabetes.
4. **FFPG (Final Fasting Plasma Glucose)**
  - Definition: Final Fasting Plasma Glucose is similar to FPG but is typically measured after several assessments, sometimes used to confirm consistent glucose levels in research or clinical settings.
  - Significance: FFPG can provide a more stable measurement of fasting glucose levels, reducing the effect of single, potentially anomalous readings.
5. **Cholesterol**
  - Definition: Cholesterol is a type of fat (lipid) found in the blood. The body needs cholesterol to build healthy cells, but high levels of cholesterol can lead to plaque buildup in arteries.
  - Significance: High cholesterol levels are associated with an increased risk of heart disease and stroke.
6. **Triglyceride**
  - Definition: Triglycerides are another type of fat in the blood. After eating, the body converts excess calories into triglycerides, which are stored in fat cells.
  - Significance: High triglyceride levels, especially in combination with high cholesterol, increase the risk of heart disease and metabolic syndrome.
7. **HDL (High-Density Lipoprotein)**
  - Definition: HDL, or "good" cholesterol, helps remove other forms of cholesterol from the bloodstream by transporting it to the liver for processing.
  - Significance: Higher levels of HDL are generally beneficial, as they help reduce the risk of heart disease by lowering "bad" cholesterol (LDL) levels.
8. **LDL (Low-Density Lipoprotein)**
  - Definition: LDL, or "bad" cholesterol, transports cholesterol particles through the body, but too much LDL can lead to plaque buildup in arteries.
  - Significance: High levels of LDL increase the risk of atherosclerosis (narrowing of arteries) and cardiovascular diseases.
9. **ALT (Alanine Aminotransferase)**
  - Definition: ALT is an enzyme found mainly in the liver. It’s released into the bloodstream when liver cells are damaged.
  - Significance: High ALT levels can indicate liver damage or disease, such as hepatitis or fatty liver disease. ALT is often checked as part of liver function tests.
10. **BUN (Blood Urea Nitrogen)**
  - Definition: BUN measures the amount of nitrogen in the blood that comes from urea, a waste product produced by the liver and removed by the kidneys.
  - Significance: High BUN levels can indicate kidney dysfunction, dehydration, or increased protein intake. Low BUN levels are less common but can be associated with malnutrition or liver disease.
11. **CCR (Creatinine Clearance)**
  - Definition: Creatinine Clearance measures how effectively the kidneys filter creatinine, a waste product from muscle metabolism, from the blood.
  - Significance: CCR is a common test for kidney function. Low CCR values suggest reduced kidney function, which can occur in chronic kidney disease or kidney failure.
12. **Smoking Status**: (1: Current Smoker, 2: Ever Smoker, 3: Never Smoker)
13. **Drinking Status**: (1: Current Drinker, 2: Ever Drinker, 3: Never Drinker)
14. **Family History of Diabetes**: (1: Yes, 0: No)

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Diabetes', data=diabetes_data)
plt.title('Class Distribution of Diabetes')
plt.xlabel('Diabetes (0: No, 1: Yes)')
plt.ylabel('Count')
plt.show()
```

```python
correlation_matrix = diabetes_data.corr()
diabetes_correlation = correlation_matrix['Diabetes'].sort_values(ascending=False)

plt.figure(figsize=(10, 6))
diabetes_correlation.drop('Diabetes').plot(kind='bar', color='skyblue')
plt.title('Correlation of Each Feature with Diabetes')
plt.xlabel('Features')
plt.ylabel('Correlation with Diabetes')
plt.show()
```
